```python
import numpy as np
#     0|2|4              model|alphabet|samples
#                           v       v       v
#      1|3                challenge|characters
#                               v       v
total_successes = np.zeros((3,  3,  5,  5,  2))
total_lengths =   np.zeros((3,  3,  5,  5,  2))
total_accuracies =   np.zeros((3,  3,  5,  5,  2))
```

```python
import glob
models = ["cnn_1", "cnn_2", "higsfa"]
for id0, model in enumerate(models):
    model_root = "../experimental_data" + "/" + model
    print(sorted(glob.glob(model_root + "/*")))
    for id1, challengeno in enumerate(sorted(glob.glob(model_root + "/*"))):    
        for id2, alphabetno in enumerate(sorted(glob.glob(challengeno + "/*"))):    
            for id3, characterno in enumerate(sorted(glob.glob(alphabetno + "/*"))):
                for id4, sampleno in enumerate(sorted(glob.glob(characterno + "/*"))):
                    print(sampleno)
                    total_success = 0
                    total_length = 0
                    accuracies = []
                    for instance in glob.iglob(sampleno + "/*"):
                        with open(instance) as f:
                            content = f.readlines()
                            success = 0
                            for line in content:
                                split = line.split(",")
                                if split[0] == split[1].strip():
                                    success = success + 1
                            total_success +=  success
                            total_length  += len(content)-1
                            accuracies.append(float(success)/len(content)-1)
                    print("hmm")
                    print(np.std(accuracies) / np.sqrt(len(accuracies)))
                    print(total_success)
                    print("ids %i, %i, %i, %i, %i" % (id0, id1, id2, id3, id4))
                    total_successes[id0, id1, id2, id3, id4] += total_success
                    total_lengths  [id0, id1, id2, id3, id4] += total_length
                    total_accuracies  [id0, id1, id2, id3, id4] += np.std(accuracies) / np.sqrt(len(accuracies))
                    print(total_success / float(total_length))

```

```python
np.save('total_successes', total_successes)
np.save('total_lengths', total_lengths)
np.save('total_accuracies', total_accuracies)
```
