```python
import glob
partitions = ["training", "testing"]
models = ["higsfa", "cnn_1", "cnn_2"]
sample_lengths = [5, 10, 50, 200, 500, 2000, 4000]

#                                model   avg/std
#                                   v
#                                    sample  partition
shape = (3, 7, 2, 2)
mnist_aggregate_results = np.zeros(shape)

model = "cnn_2"



id0 = 0
print(sorted(glob.glob(model_root + "/*")))
print("\n")
for model in models:
    model_root = "../experimental_data_mnist" + "/" + model
    for id1, samplesno in enumerate(sorted(glob.glob(model_root + "/*"))):    
        for id2, data_partition in enumerate(sorted(glob.glob(samplesno + "/*"))):    
            print(data_partition)
            sample_length = int(data_partition.split("/")[3].split("_")[0])
            print(data_partition.split("/")[2])
            print(data_partition.split("/")[4])
            print(sample_length)
            total_success = 0
            total_length = 0     
            pct = []
            for id3, runno in enumerate(sorted(glob.glob(data_partition + "/*"))):
                with open(runno) as f:
                    content = f.readlines()
                    success = 0                
                    for line in content:
                        split = line.split(",")
                        if split[1].strip() == split[2].strip():
                            success = success + 1
                    total_success +=  success
                    total_length  += len(content)-1
                    pct.append(success/float(len(content)-1))
            print(np.std(pct) / np.sqrt(10.0))
            if id2 == 1:
                #print("total training samples: total_length / 100 = %i" %(total_length/100))
                print("classification error: %f" % (total_success / float(total_length)) )
            else:
                #print("test samples per iteration: %i" %(total_length/100))
                print("classification error: %f" % (total_success / float(total_length)) )
            print("ids %i, %i, %i " % (id0, id1, id2))

            sample_id = False
            models_id = False
            partition_id = False
            for i in range(7):
                if sample_length == int(sample_lengths[i]):
                    sample_id = i
            for j in range(3):
                if str(data_partition.split("/")[2]) == str(models[j]):
                    models_id = j
            for k in range(2):
                #if str(data_partition.split("/")[4]) == str(partitions[k]):
                if partitions[k] in data_partition:
                    partition_id = k
            #print(mnist_aggregate_results)
            mnist_aggregate_results[models_id, sample_id, 0, partition_id] = (total_success / float(total_length))
            mnist_aggregate_results[models_id, sample_id, 1, partition_id] = np.std(pct) / np.sqrt(10.0)



            total_successes[id0, id1, id2] += total_success
            total_lengths  [id0, id1, id2] += total_length
            print("\n")
# higsfa, training avg:
print(",".join(map(str, mnist_aggregate_results[0, :, 0, 0])))
# test
print(",".join(map(str, mnist_aggregate_results[0, :, 0, 1])))
# training std
print(",".join(map(str, mnist_aggregate_results[0, :, 1, 0])))
# test
print(",".join(map(str, mnist_aggregate_results[0, :, 1, 1])))
```
